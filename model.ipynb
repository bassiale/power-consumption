```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
#ritrasformo i dati in un dataframe
data = pd.read_csv('/data.csv')
data.head()
```

```python
#separo i dati per l'allenamento del modello e i risultati
train_x = data.loc[:,'membri_familiari':'condizionatore']
train_y = data.loc[:, '1F1':]
```

```python
#creo un modello(random forest)
forest_model = RandomForestRegressor(random_state=1)
#lo alleno
forest_model.fit(train_x, train_y)
#calcolo l'errore sul dataset
melb_preds = forest_model.predict(train_x)
print(mean_absolute_error(train_y, melb_preds))
```
